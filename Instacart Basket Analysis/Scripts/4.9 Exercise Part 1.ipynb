```python
# Load the orders-products merged data
ords_prods_merge = pd.read_pickle(os.path.join(path, 'Data', 'Prepared Data', 'ords_prods_merge_4.9_.pkl'))
```

#######'user_id' is the key column to comnbine both dataframes

```python
#Drop the '_merge' column from the ords_prod_merge dataframe
if '_merge' in ords_prods_merge.columns:
    ords_prods_merge = ords_prods_merge.drop('_merge', axis=1)
```

```python
#check the output 
ords_prods_merge.columns
```

```python
ords_prods_merge.shape
```

```python
#Merge (join) ords_prods_merge with customers data frame on user_id

ords_prods_merge_custs = ords_prods_merge.merge(
    df_custs, 
    on='user_id', 
    how='inner', 
    suffixes=('', '_cust')  # add a suffix for overlapping columns if needed
)
```
