```python
user_avg_price = ords_prods_merge.groupby('user_id')['prices'].mean().reset_index()
```

#######We first group the data by user_id and calculate the mean price of products purchased by each user.

```python
def spending_flag(avg_price):
    if avg_price < 10:
        return "Low spender"
    else:
        return "High spender"
```

Users with average price < 10 â†’ "Low spender"

Users with average price â‰¥ 10 â†’ "High spender"

```python
user_avg_price.rename(columns={'prices': 'avg_price'}, inplace=True)
```

#######Rename the prices column to avg_price for easier understanding.

```python
user_avg_price['spending_flag'] = user_avg_price['avg_price'].apply(spending_flag)
```

#######Create a new column spending_flag by applying the function to the avg_price column.

```python
ords_prods_merge = ords_prods_merge.merge(user_avg_price[['user_id', 'spending_flag']], on='user_id', how='left'
)
```

########Every row in ords_prods_merge now has a spending_flag column

```python
print(ords_prods_merge['spending_flag'].value_counts())
```

########This shows how many users are Low spenders vs High spenders.The majority of users (â‰ˆ 29.8 million orders) fall into the Low spender category.Only a small fraction of users (â‰ˆ 0.6 million orders) are High spenders.
This indicates that most customers purchase lower-priced items, while a small group contributes to higher-value sales.

```python
user_order_median = ords_prods_merge.groupby('user_id')['days_since_prior_order'].median().reset_index()
```

#######We first calculate the median number of days between orders for each user. This gives a better picture of their typical ordering behavior than the mean, because it is robust to outliers.

```python
def order_frequency_flag(median_days):
    if median_days > 20:
        return "Non-frequent customer"
    elif median_days > 10 and median_days <= 20:
        return "Regular customer"
    else:
        return "Frequent customer"
```

The flag is defined using the following criteria: median_days > 20 â†’ "Non-frequent customer" median_days > 10 and <= 20 â†’ "Regular customer" median_days <= 10 â†’ "Frequent customer"

```python
user_order_median['order_frequency_flag'] = user_order_median['days_since_prior_order'].apply(order_frequency_flag)
```

########Every row in ords_prods_merge now contains the order_frequency_flag.

```python
ords_prods_merge = ords_prods_merge.merge(user_order_median[['user_id', 'order_frequency_flag']],on='user_id', how='left'
)
```

```python
print(ords_prods_merge['order_frequency_flag'].value_counts())
```

########This shows how many users are Frequent, Regular, or Non-frequent customers.
