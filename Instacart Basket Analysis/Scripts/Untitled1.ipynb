```python
#Importing Libraries 
import pandas as pd
import numpy as np
import os 

path = r'C:\Users\ANITA BOADU\Achievement 4'

# Import the dataframe ords_prods_merge as 
ords_prods_merge = pd.read_pickle(os.path.join(path,'Data','Prepared Data','ords_prods_merge.pkl'))

# Create the first 1,000,000-row subset
df = ords_prods_merge[:1000000]
```

```python
df.groupby('product_name')
```

########Creates a grouped object where all rows with the same product_name
are grouped together.

By itself, this command does not produce a visible result.

It prepares the data so you can apply operations like counting, summing, or aggregating.

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

#######Groups rows based on department_id and applies the mean function to the order_number column

Returns a table showing the average order number for each department

```python
df.groupby('department_id')['order_number'].mean()
```

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

########Groups all rows by department_id and For each department, it calculates:Mean (average order number), Minimum order number and the Maximum order number. The result is a table showing how order numbers vary across different departments.

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('max')
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

########The numbers represent how many orders or customers fall into each loyalty category.
Most customers are regular shoppers, but a large portion are loyal, indicating a healthy returning customer base.
The new customers show ongoing growth potential.
